```python
from nilearn import plotting
import os
```

```python
ImagesDir = 'indiv_a2009s_aseg_ROIs/'
imageFiles = [file for file in os.listdir(ImagesDir)]

for i in range(len(imageFiles)):
    imageFile = imageFiles[i]
    image = ImagesDir + imageFile
    saveFile = 'nil_' + imageFile + '.png'
    display = plotting.plot_roi(image, draw_cross=False, cmap='autumn')
    display.add_contours(image, filled=True, alpha=0.7, levels=[0.5], colors='r')
    display.savefig(saveFile)
    display.close()
```
